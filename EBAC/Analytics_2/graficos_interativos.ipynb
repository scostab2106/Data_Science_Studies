```python
import pandas as pd
import numpy as np
import plotly.express as px
```

Criação de DataFrame fictício


```python
idades = np.random.randint(20, 60, 20) # Gerar 20 idades aleatórias entre 20 e 60
salarios = idades * np.random.randint(80, 120, 20) # Gerar salários baseados nas idades 
pontuacoes = salarios * np.random.uniform(0.5, 1.5, 20) # Gerar pontuações baseadas nos salários 
profissoes = ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Designer', 'Analista', 'Gerente', 'Programador'] # Lista de profissões
estado_civil = ['Solteiro', 'Casado', 'Divorciado', 'Viúvo'] # Lista de estados civis

novo_df = {
    'Idade': idades,
    'Salário': salarios,
    'Pontuação': pontuacoes.round(2) # Arredondar as pontuações para 2 casas decimais
}

df = pd.DataFrame(novo_df) # Criar um DataFrame a partir do dicionário
df['Profissão'] = np.random.choice(profissoes, size=len(df)) # Adicionar uma coluna de profissões aleatórias
df['Estado Civil'] = np.random.choice(estado_civil, size=len(df)) # Adicionar uma coluna de estados civis aleatórios
df.head(10) # Exibir as 10 primeiras linhas do DataFrame
```

Plotando gráfico de barras horizontais com a média dos salários por profissão

```python
#Agrupando e calculando a média salarial por profissão
salario_por_profissao = df.groupby('Profissão')['Salário'].mean().reset_index() # Agrupar por profissão e calcular a média salarial'
#Criando gráfico de barras horizontais
fig = px.bar(salario_por_profissao, x='Salário', y='Profissão', orientation='h', title='Salários por profissão', labels={'Salário': 'Salário Médio', 'Profissão': 'Profissão'})
#Exibir o gráfico
fig.show()
```

Com mapa de calor

```python
#Agrupando e calculando a média salarial por profissão
salario_por_profissao = df.groupby('Profissão')['Salário'].mean().reset_index() # Agrupar por profissão e calcular a média salarial'
#Criando gráfico de barras horizontais
fig = px.bar(salario_por_profissao, 
             x='Salário', 
             y='Profissão', 
             orientation='h', 
             title='Salários por profissão', 
             labels={'Salário': 'Salário Médio', 'Profissão': 'Profissão'},
             color='Salário', # Adicionando a cor baseada nos valores de salário
             width=800) # Definindo a largura do gráfico
#Exibir o gráfico
fig.show()
```

Mapa de árvore

```python
salario_por_profissao_estado_civil = df.groupby(['Profissão', 'Estado Civil'])['Salário'].mean().reset_index() # Agrupar por profissão e calcular a média salarial'
fig = px.treemap(salario_por_profissao_estado_civil, 
                 path=['Profissão', 'Estado Civil'], 
                 values='Salário', 
                 title='Salários por profissão e estado civil', 
                 color='Estado Civil')
fig.show()
```
