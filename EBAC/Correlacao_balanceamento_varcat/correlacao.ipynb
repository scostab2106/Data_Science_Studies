```python
correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 10}) # annot=True para mostrar os valores, fmt='.2f' para formatar os números com 2 casas decimais e annot_kws para ajustar o tamanho da fonte dos valores
plt.title('Matriz de Correlação')
plt.show()
```
