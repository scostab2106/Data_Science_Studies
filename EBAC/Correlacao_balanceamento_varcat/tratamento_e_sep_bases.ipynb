```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:\\Users\\samco\\OneDrive\\LP\\Repositorios DEV\\EBAC_Estudos\\EBAC\\Pre_processamento_dados\\CHURN_CREDIT_MOD14.csv', delimiter=',')

df.head(10)
```

```python
# Critando uma instancia do LabelEncoder
label_encoder = LabelEncoder()

# Aplicando para a coluna "Genero"
df['Genero_encoded'] = label_encoder.fit_transform(df['Genero'])

# Aplicar o Onehot para a coluna "Pais"
df = pd.get_dummies(df, columns=['Pais'], prefix='Pais', drop_first=True) # o drop first=True é para evitar a multicolinearidade, ou seja, evitar que uma coluna seja completamente explicada por outras colunas, nesse caso, quando não for spain ou germany o valor será france, então não é necessário criar uma coluna para isso

print(df.head())
```

TRANSFORMANDO AS COLUNAS DOS PAÍSES EM INTEIRO

```python
for column in df.columns:
    if df[column].dtype == 'bool':
        df[column] = df[column].astype(int)
print(df.head())
```

```python
# Dropando a coluna "Genero" original pois ainda é categórica
df.drop('Genero', axis=1, inplace=True)
```

SEPARAÇÃO ENTRE BASE TREINO E TESTE

```python
# Separar os dados em features (X) e target (y)
X = df.drop('Churn', axis=1) # Considerando que Churn é a coluna a ser predita
y = df['Churn'] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Separar os dados em conjuntos treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
# O parâmetro test_size=0.25 indica que 25% dos dados serão usados para teste e 75% para treino. O random_state=42 é usado para garantir a reprodutibilidade dos resultados, ou seja, sempre que o código for executado, a divisão dos dados será a mesma.
```

BALANCEAMENTO

```python
# Calcular e imprimir as porcentagens dos valores na coluna 'churn'ArithmeticError

print((df['Churn'].value_counts(normalize=True) * 100))
```

```python
from imblearn.over_sampling import SMOTE

# Criar uma instancia do SMOTE
smote = SMOTE(random_state=42)

# Aplicar o SMOTE aos dados de treinamento (x_train e y_train)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Verificar a distribuição das classes após o balanceamento
print('Distribuição das classes após o balanceamento:')
print(y_train_balanced.value_counts())
```

```python
# Salvando as bases

y_train_balanced.to_csv('y_train_balanced.csv', index=False)
X_train_balanced.to_csv('X_train_balanced.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
```
