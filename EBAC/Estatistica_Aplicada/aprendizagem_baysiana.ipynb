```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
X_test = pd.read_csv('X_test.csv', delimiter=',')
X_train = pd.read_csv('X_train_balanced.csv', delimiter=',')
y_test = pd.read_csv('y_test.csv', delimiter=',')
y_train = pd.read_csv('y_train_balanced.csv', delimiter=',')
```

Verificar o balanceamento

```python
# Contar o número de exemplos de cada classe em y_trrain
train_balance = y_train.value_counts()
print("Balanceamento em y_train:")
print(train_balance)

# Contar o número de exemplos de cada classe em y_test
test_balance = y_test.value_counts()
print("\nBalanceamento em y_test:")
print(test_balance)
```

    O certo é entregar a base de teste desbalanceada para o modelo

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
# Criação de uma instancia do classificador Naive Bayes Gaussiano
naive_churn = GaussianNB()

# Treinamento do classificador usando os dados de treinamento
naive_churn.fit(X_train, y_train)
```

```python
# Fazer previsões para os dados de treino
y_pred_train = naive_churn.predict(X_train)

# Calcular a acuracia
accuracy = accuracy_score(y_train, y_pred_train)
print(f"Acurácia do modelo nos dados de treino:" , accuracy)
```

Criando gráfico da matriz de confusão

```python



import plotly.figure_factory as ff


conf_matrix = confusion_matrix(y_train, y_pred_train)

# Definindo nomes de classes
class_names = ['Churn', 'Não Churn']

# Plotando a matriz de confusão com Plotly
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Viridis',
    showscale=True
)

fig.update_layout(
    title='Matriz de Confusão - Treinamento',
    xaxis_title='Classe Predita',
    yaxis_title='Classe Verdadeira',
    font=dict(
        family="Arial",
        size=12,
        color="black"
    )
    
)
```
