EXEMPLO 2 - CIDADES

Problemas de acentuação e letras maiúsculas e minúsculas costuma diminuir a qualidade dos dados

```python
dados = {
    'Idade': [25, 30, 35, 40, 45, 400, 554, 233, 50, 60, 60, 70, 26],
    'Cidade' : ['São Paulo', 'Curitiba', 'Salvador', 'Cuitiba', 'Sao Paulo', 'salvador', 'salvador', 'sao paulo', 'curitiba', 'Curitiba', 'Sao Paulo', 'Salvado', 'salvador']
}

df = pd.DataFrame(dados)
print(df)
```

Média das idades por cidade realizando agrupamento por cidades:

```python
df.groupby('Cidade')['Idade'].mean().reset_index

#Por causa dos erros são mostrados vários valores como se tivessem muitas cidades, sendo que existem apenas 3 nos dados.
```
