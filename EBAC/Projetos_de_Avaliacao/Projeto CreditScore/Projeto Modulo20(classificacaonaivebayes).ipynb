```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

Durante a aula nossa variável a ser prevista (churn) continha apenas 2 categorias, a base de vocês contém mais. O Naive Bayes pode ser aplicado para problemas de classificação com múltiplas classes da mesma forma que para problemas de classificação binária. O Naive Bayes é um algoritmo de classificação probabilístico que calcula a probabilidade de uma amostra pertencer a cada classe e seleciona a classe com a maior probabilidade como a previsão final.
Em resumo, o Naive Bayes pode ser aplicado da mesma maneira para problemas de classificação com múltiplas classes, e os mesmos princípios se aplicam em termos de treinamento, avaliação e aplicação do modelo.

```python
#seu código aqui

X_test = pd.read_csv('X_test_balanced.csv')
y_test = pd.read_csv('y_test_balanced.csv')
X_train = pd.read_csv('X_train_balanced.csv')
y_train = pd.read_csv('y_train_balanced.csv')
```

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

```python
#seu código aqui

naive = GaussianNB()

naive.fit(X_train, y_train)
```

```python
y_train_pred = naive.predict(X_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
#seu código aqui

accuracy = accuracy_score(y_train, y_train_pred)
recall = recall_score(y_train, y_train_pred, average='macro')

print('Acurácia no conjunto de treino: ', accuracy)
print('Recall no conjunto de treino: ', recall)
```

```python
import plotly.figure_factory as ff


conf_matrix = confusion_matrix(y_train, y_train_pred)

# Definindo nomes de classes
class_names = ['0', '1', '2']

# Plotando a matriz de confusão com Plotly
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Viridis',
    showscale=True
)

fig.update_layout(
    title='Matriz de Confusão - Treinamento',
    xaxis_title='Classe Predita',
    yaxis_title='Classe Verdadeira',
    font=dict(
        family="Arial",
        size=12,
        color="black"
    )
    
)
```

```python
# O modelo foi muito bem, com mais de 98% de acurácia e recall e apenasa 4 erros de classificação. Me chamou atenção os valores quase identicos de acurácia e recall.
```

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados, da mesma forma que fez acima. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
y_test
```

```python
#seu código aqui

naive.fit(X_test, y_test)

y_test_pred = naive.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_test_pred)
recall = recall_score(y_test, y_test_pred, average='macro')

print('Acurácia no conjunto de teste: ', accuracy)
print('Recall no conjunto de teste: ', recall)
```

```python

conf_matrix = confusion_matrix(y_test, y_test_pred)

# Definindo nomes de classes
class_names = ['0', '1', '2']

# Plotando a matriz de confusão com Plotly
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=class_names,
    y=class_names,
    colorscale='Viridis',
    showscale=True
)

fig.update_layout(
    title='Matriz de Confusão - Teste',
    xaxis_title='Classe Predita',
    yaxis_title='Classe Verdadeira',
    font=dict(
        family="Arial",
        size=12,
        color="black"
    )
    
)
```

```python
# O modelo lidando com a base teste teve desempenho perfeito alcançando 100% de acurácia e recall, sem nenhum erro de classificação. Isso indica que o modelo generalizou muito bem para novos dados, o que é um resultado excelente. Porém essa perfeição pode vir tamebém da pequena quantidade previsões na base de teste.
```
