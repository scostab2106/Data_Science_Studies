*   **Tempo_como_Cliente** : Tempo em meses que nosso cliente está com a nossa empresa.

*   **PhoneService** : Se ele contratou serviço telefônico com a nossa empresa.

*   **Servico_Internet** : Qual o tipo de serviço de internet contratado.

*   **Servico_Seguranca** : Se ele contratou serviço de segurança com a nossa empresa.

*   **Suporte_Tecnico** : Se já acionou o suporte técnico.

*   **StreamingTV** : Se ele contratou serviço de streaming para TV.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **PaymentMethod** : Forma de Pagamento.

*   **Pagamento_Mensal** : Quanto o cliente pagava mensalmente.

*   **Total_Pago** : Total já pago pelo cliente para nossa companhia.

*   **Churn** : Nossa coluna mais importante, se nosso cliente nos abandonou ou não.




# 1 - Carregue a base, verifique se os tipos de dados estão de acordo com cada coluna e caso não estejam realize a transformação dos tipos de dados.

```python
#seu código aqui

df.dtypes
```

```python
df['Churn'] = df['Churn'].replace({'No': 0, 'Yes': 1})
df['Casado'] = df['Casado'].replace({'No': 0, 'Yes': 1})
df['Dependents'] = df['Dependents'].replace({'No': 0, 'Yes': 1})
df['Servico_Seguranca'] = df['Servico_Seguranca'].replace({'No': 0, 'Yes': 1, 'No internet service': 0})
df['Suporte_Tecnico'] = df['Suporte_Tecnico'].replace({'No': 0, 'Yes': 1, 'No internet service': 0})
df['StreamingTV'] = df['StreamingTV'].replace({'No': 0, 'Yes': 1, 'No internet service': 0})


df['Casado'] = df['Casado'].astype(int)
df['Dependents'] = df['Dependents'].astype(int)
df['Servico_Seguranca'] = df['Servico_Seguranca'].astype(int)
df['Suporte_Tecnico'] = df['Suporte_Tecnico'].astype(int)
df['StreamingTV'] = df['StreamingTV'].astype(int)
```

```python
df_churn_nulo = df[df['Churn'].isnull()]
df_churn_nulo.head(10)

# Como podemos verificar aqui, os dados nulos da coluna churn são uma parte muito pequena dos dados da coluna, e considerando as outras colunas, podemos observar que os nulos não se limitam a um tipo específico de cliente, ou seja, os dados nulos estão distribuídos entre os clientes, então remover os dados nulos é a melhor opção pois não existe um risco de enviesar os dados.
```

```python
#seu código aqui

df.dropna(subset=['Genero'],inplace=True)
df.dropna(subset=['Churn'],inplace=True)

# A coluna gênero possuia apenas 0,4% de seus dados como nulos, por não representar parte relevante dos dados da coluna e a coluna Genero não ser tão relevante para o modelo, optei por retirar os dados nulos dessa coluna.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='PhoneService')
plt.title('Contagem de Serviços de Telefone')
plt.xlabel('Serviço de Telefone')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()


df['PhoneService'].value_counts(normalize = True) * 100 
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Churn', hue='PhoneService')
plt.title('Contagem de Churn por Serviço de Telefone')
plt.xlabel('Churn')
plt.ylabel('PhoneService')
plt.show()

# Analisando este gráfico é possível perceber que não existe muita correlação entre a coluna PhoneService e a coluna Churn, ou seja, o fato do cliente possuir ou não serviço de telefone não é um fator relevante para o modelo prever se o cliente irá ou não cancelar o serviço, então essa coluna que possui mais da metade de dados nulos pode ser removida.
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Pagamento_Mensal', color='skyblue')
plt.title('Boxplot do Pagamento Mensal')
plt.ylabel('Pagamento Mensal')
plt.show()

df['Pagamento_Mensal'].median()
#df['Pagamento_Mensal'].mean()

# Observando a media e a mediana da coluna Pagamento_Mensal, podemos perceber o valor da média um pouco menor que o valor da mediana, porém ao verificar o boxplot não encontramos outliers, apenas uma grande diferença entre os valores de pagamento mensal, então optei por preencher os dados nulos com a média da coluna, pois a média é um valor mais representativo para os dados da coluna do que a mediana.
```

```python
#seu código aqui

df.drop(columns=['PhoneService'], inplace=True)
df['Pagamento_Mensal'].fillna(df['Pagamento_Mensal'].mean(), inplace=True)
```

```python
df.drop(columns=['customerID'], inplace=True)
```
