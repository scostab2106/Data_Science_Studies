```python
df_grouped = df.groupby(['Churn', 'Casado']).size().reset_index(name='count')
total_por_churn = df_grouped.groupby('Churn')['count'].transform('sum')
df_grouped['percent'] = (df_grouped['count'] / total_por_churn) * 100

# Criar o gráfico de barras empilhadas
fig = px.bar(df_grouped, x='Churn', y='percent', color='Casado', labels={'percent': 'Porcentagem', 'Churn': 'Churn', 'Casado': 'Casado'})

# Layout
fig.update_layout(
    title='Distribuição de Churn por Casado',
    yaxis_title='Porcentagem',
    legend_title='Casado'
)
fig.show()

# Dos clientes que deram churn, temos mais solteiros do que casados, mas a diferença não é tão grande, então acredito que o fato de ser casado ou não não seja um fator tão relevante para o churn dos clientes
```

```python

df_grouped = df.groupby(['Churn', 'Genero']).size().reset_index(name='count')
total_por_churn = df_grouped.groupby('Churn')['count'].transform('sum')
df_grouped['percent'] = (df_grouped['count'] / total_por_churn) * 100

# Criar o gráfico de barras empilhadas
fig = px.bar(df_grouped, x='Churn', y='percent', color='Genero', labels={'percent': 'Porcentagem', 'Churn': 'Churn', 'Genero': 'Gênero'})

# Layout
fig.update_layout(
    title='Distribuição de Churn por Gênero',
    yaxis_title='Porcentagem',
    legend_title='Gênero'
)
fig.show()

# Vemos um núemero bem equilibrado de proporção de churn entre os gêneros, então acredito que o gênero dos clientes não seja um fator tão relevante para o churn dos clientes.
```

Conforme a Análise realizada, foi possível notar variáveis bem relevantes ao contexto do churn:

Clientes que tem o contrato mensal estão mais propensos ao abandono do serviço, a questão é que um contrato mensal relmente facilita o cliente abandonar o serviço por não ter que pagar nenhuma multa e ter a aportunidade de trocar de serviço quando quiser.

Clientes que coontratam o serviço de internet de fibra óptica estão dando mais churn do que os que contratam DSL ou os que não contratam, mesmo que fiquem a mesma média de tempo com o serviço. Isso mostra uma insatisfação com o serviço de fibra óptica ofertado demonstrando que o serviço precisa melhorias.

Cliente que utilizam como meio de pagamento o check eletronico costumam dar mais churn que os outros meios de pagamento provavelmente por conta da burocracia ou alguma outra dificuldade, extinguir essa alternativa de meio de pagamento poderia ser uma opção, pois ela não chega a ser disparadamente mais utilizada que as outras.

As pessoas que tem mais tempo como clientes estão dispostos a pagar mais pelos serviços e dão menos churn mostrando que adiquiriram confiança na empresa.
