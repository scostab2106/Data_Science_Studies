```python
import pandas as pd
```

```python
dados = {
    'Investimento': [100, 140, 160, 390, 380, 245, 250, 140, 360],
    'impressao': [2050, 2500, 3350, 7500, 7120, 3650, 1985, 2150, 2100],
    'Alcance': [1500, 2350, 2950, 7200, 6558, 2050, 1185, 1930, 1985],
    'Curtidas': [200, 20, 100, 120, 125, 125, 90, 50, 55],
    'Comentarios': [50, 15, 10, 45, 56, 60, 2, 15, 32],
    'Acessos': [200, 260, 320, 780, 700, 340, 420, 180, 200]
}

df = pd.DataFrame(dados)
```

PASSO 1 - MATRIZ DE CORRELAÇÃO

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
correlacao = df.corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
```

PASSO 2 - SEPARAÇÃO DE VARIÁVELDEPENDENTE E INDEPENDENTES.

```python
X = df[['Investimento', 'impressao', 'Alcance', 'Curtidas', 'Comentarios']] # Independentes
y = df['Acessos'] # Dependente(RÓTULO)  
```

PASSO 3 - REGRESSÃO LINEAR MÚLTIPLA

```python
from sklearn.linear_model import LinearRegression   
```

```python
regressao_multipla = LinearRegression()
regressao_multipla.fit(X, y)
```

```python
LinearRegression()
```

```python
regressao_multipla.intercept_
```

```python
regressao_multipla.coef_
```

PASSO 4 - AVALIAÇÃO DO MODELO   

```python
regressao_multipla.score(X, y)
```
