```python
import pandas as pd
```

```python
# Dados de investimento em publicidade e acessos no site

dados = {
    'Investimento': [100, 140, 150, 390, 350, 155, 200],
    'Acessos': [200, 260, 300, 760, 700, 320, 400]
}

df = pd.DataFrame(dados)

```

```python
# Calcular correlação

df['Investimento'].corr(df['Acessos'])
```

Regressão linear simples com Scikit-Learn

```python
from sklearn.linear_model import LinearRegression

# PASSO 1: Separar as variáveis independentes (X) e dependentes (y)

X = df[['Investimento']]  # Variável independente(características)
y = df['Acessos']  # Variável dependente (rótulo)

```

```python
X
```

```python
# PASSO 2: Criar o modelo de regressão

# Esta linha cria um objeto do tipo LinearRegression, que será usado para representar nosso modelo de regressão linear.
regressao_acessos_site = LinearRegression()

regressao_acessos_site.fit(X, y)
# O modelo aprende a relação entre as variáveis independentes e a variável dependente.
# Após a execução dessa linha o modelo estará pronto para fazer previsões.
```

```python
# O regressao_acessos_site.intercept_ é um atributo do objeto de regressão linear que representa o coeficiente linear da equação
regressao_acessos_site.intercept_
```

```python
# O regressao_acessos_site.coef é um atributo do objeto de regressão linear que representa os coeficientes das variáveis independentes na equação de regressão linear
regressao_acessos_site.coef_
```

```python
# EQUAÇÃO VAI SER: Acessos = 3.2*(Investimento) + 1.96
```

```python
# Passo 3: Calcular o R² do modelo

regressao_acessos_site.score(X,y)
```

```python
import matplotlib.pyplot as plt
```

```python
# Passo 4: Visualizar os dados e a linha de regressão

# Plotar os dados originais

plt.scatter(X,y, color='blue', label='Dados Originais')

# Plotar a linha de regressão
plt.plot(X, regressao_acessos_site.predict(X), color='red', label='Linha de Regressão')
plt.legend()
plt.show()
```

```python
# Passo 5: Realizando previsões
previsoes = regressao_acessos_site.predict(X)
previsoes
```
