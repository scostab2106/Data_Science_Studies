```python
import pandas as pd
```

```python
dados = {
    'Investimento': [100, 140, 160, 390, 380, 245, 250, 140, 360],
    'impressao': [2050, 2500, 3350, 7500, 7120, 3650, 1985, 2150, 2100],
    'Alcance': [1500, 2350, 2950, 7200, 6558, 2050, 1185, 1930, 1985],
    'Curtidas': [200, 20, 100, 120, 125, 125, 90, 50, 55],
    'Comentarios': [50, 15, 10, 45, 56, 60, 2, 15, 32],
    'Acessos': [200, 260, 320, 780, 700, 340, 420, 180, 200]
}

df = pd.DataFrame(dados)
```

```python
X = df[['Investimento', 'impressao', 'Alcance', 'Curtidas', 'Comentarios']] # Independentes
y = df['Acessos'] # Dependente(RÓTULO)  
```

```python
import statsmodels.api as sm
```

```python
# Adiciona a constante (intercepto) ao conjunto de dados para estimar o termo de interceptação
X = sm.add_constant(X) 

# Criando o modelo de regressão múltipla --- (Ordinary Least Squares, Mínimos Quadrados Ordinários)
modelo_stats = sm.OLS(y, X)

# Ajustando o modelo aos dados
resultado_stats = modelo_stats.fit()

# Exibir os resultados da regressão
print(resultado_stats.summary())
```
